### Netoyage des données

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('data.xlsx').drop(['DP', 'Obs'], axis=1)
df
```

```python
#get na count for each column descending
df.isna().sum().sort_values(ascending=False)
#show full list
```

```python
#drop colum with more than 60% na
df = df.dropna(thresh=len(df)*0.5, axis=1)
df
```

```python
def plot2D(X, y, col=None):
    if col == None:
        sns.pairplot(X.join(y), hue="color")
    else:
        sns.pairplot(X[col].join(y), hue="color")
```

```python
col = ['y0_RGB', 'ymax_RGB', 'k_RGB', 'lag_RGB']
x = df.drop('color', axis=1)
y = df[['color']]
plot2D(x, y, col)
```

### Sans normalisation des données

```python
# replace missing values using decision tree
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.tree import DecisionTreeRegressor

# impute missing numeric values
df_numeric = df.select_dtypes(include=np.number)
df_numeric = pd.DataFrame(IterativeImputer(DecisionTreeRegressor()).fit_transform(df_numeric), columns=df_numeric.columns)
```

```python
#add missing categorical values
df_categorical = df.select_dtypes(exclude=np.number)
df2 = pd.concat([df_numeric, df_categorical], axis=1).drop(['Camera1', 'num_plt'], axis=1)
df2
```

```python
from sklearn.preprocessing import StandardScaler

# seperate features and target

x = df2.drop('color', axis=1)
# Standardizing the features
#x = StandardScaler().fit_transform(x)
#x = pd.DataFrame(x, columns=df2.drop('color', axis=1).columns)

y = df2['color']
```

```python
x
```

```python
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf
```

```python
finalDf = pd.concat([principalDf, df2[['color']]], axis = 1)
finalDf
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Green', 'Red']
colors = ['g', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['color'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

### Netoyage outliers avec quantile

```python
Q1 = df2.quantile(0.05)
Q3 = df2.quantile(0.95)
IQR = Q3 - Q1

df3 = df2[~((df2 < (Q1 - 1.5 * IQR)) |(df2 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df3
```

```python
x = df3.drop('color', axis=1)
# Standardizing the features
x = StandardScaler().fit_transform(x)
x = pd.DataFrame(x, columns=df2.drop('color', axis=1).columns)

y = df3['color']
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf, df3[['color']]], axis = 1)

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Green', 'Red']
colors = ['g', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['color'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

### Netoyage vif

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
 
X=df2.drop(["color"],axis=1)

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
 
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
 
print(vif_data)
```

```python
liste_drop=[vif_data["feature"][i] for i in range(len(vif_data)) if vif_data["VIF"][i] >5]
dfbis=df2.drop(liste_drop,axis=1)
dfbis
```
